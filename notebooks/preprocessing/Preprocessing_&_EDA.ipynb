# Data Preprocessing

```python
path = '/content/drive/MyDrive/CVR College/Mini Project/data/Finalized data/extracted_dataset.csv'
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load and Inspect

```python
# Load
df = pd.read_csv(path)
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.shape
```

```python
df.columns
```

```python
# Rename columns
df.columns = df.columns.str.replace(r'^[yx]\d*_', '', regex=True)

# New columns
print(df.columns.tolist())
```

## EDA

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Drops any row containing at least one null value
df = df.dropna()
```

```python
# Plot distribution of teu count
plt.figure(figsize=(10, 6))
sns.histplot(df['teu_count'], bins=30, kde=True)
plt.title('Distribution of TEU Count')
plt.xlabel('TEU Count')
plt.ylabel('Frequency')
plt.show()
```
