# LightGBM Model building

```python
import lightgbm as lgb
import joblib
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
path = '/content/drive/MyDrive/CVR College/Mini Project/data/Finalized data/preprocessed_data.csv'
```

## Load

```python
df = pd.read_csv(path)
df.head()
```

```python
df.columns
```

## Feature Separation

```python
target_col = 'arrival_delay_time'

X = df.drop(columns=[target_col])
y = df[target_col]

print(f"Features (X) shape: {X.shape}")
print(f"Target (y) shape: {y.shape}")
```

## Train Test Split

```python
# Split the data: 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training samples: {X_train.shape[0]}")
print(f"Testing samples: {X_test.shape[0]}")
```

## Build Model

```python
lgbm_model = lgb.LGBMRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
lgbm_model.fit(X_train, y_train)
```

## Evaluate Model

```python
# Predict delay times on the unseen test set
y_pred_lgbm = lgbm_model.predict(X_test)

# Calculate error metrics
rmse_lgbm = np.sqrt(mean_squared_error(y_test, y_pred_lgbm))
mae_lgbm = mean_absolute_error(y_test, y_pred_lgbm)
r2_lgbm = r2_score(y_test, y_pred_lgbm)

print("--- LightGBM Evaluation ---")
print(f"Root Mean Squared Error (RMSE): {rmse_lgbm:.2f}")
print(f"Mean Absolute Error (MAE):      {mae_lgbm:.2f}")
print(f"R-squared Score (R2):           {r2_lgbm:.4f}")
```

## Export

```python
# Export model
import joblib

# Export the model to a file
joblib.dump(lgbm_model, 'lightGBM.joblib')

# --- To load it back later ---
# loaded_model = joblib.load('my_model.joblib')
```
